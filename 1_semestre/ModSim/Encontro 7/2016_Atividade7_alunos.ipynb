**(d)** O próximo passo é gerar o código que efetivamente calculará os valores das populações de alces e lobos em função do tempo. Crie um script que, utilizando DeltaL e DeltaA, determine os valores das populações em função do tempo. 
Precisaremos definir valores iniciais para a população de alces e lobos (L(0) e A(0)) além dos parâmetros do sistema e o tempo total de cálculo. 
De momento não trabalhermos com listas (vocês devem estar aprendendo isto em SofDes agora!), imprimiremos o resultado na tela, e copiaremos para o Excel para poder graficar lá. 
A continuação, devemos definir por quanto tempo queremos rodar nossa simulação. Chamaremos esta variavel de Tmax. 

```python
Tmax = 500
```

Escreva o código que calcula L(t) e A(t):

```python
# Calculando a população de alces e lobos no tempo
x = 0
lt = 20
at = 1120
while x <= Tmax:
    deltal = gama * lt * (at/alces_criticos - 1)  
    deltaA = beta * at * (1 - lt/lobos_criticos)
    lt += deltal
    at += deltaA
    x += 1
    print (at, lt) 
    
```
