**Problema 1.** Escreva uma sequência de instruções que avalie a seguinte expressão (1). Você deve usar pelo menos uma variável para armazenar o valor de uma subexpressão. Avalie a expressão para os valores de $\mu$ = -2, $\sigma$ = 0.5, e x = -1.

Note que o numerador refere-se à função exponencial.
Na resposta a este problema, por favor forneça o código comentado.

Nota: para quem esta utilizando o Chrome no Mac, ao digitar equações aparecerá uma barra vertical no final da equação. Este é um bug do Chrome no Mac, não se preocupe com isto por agora.

```python
import math 
mi = -2
sigma = 0.5
x = -1

a = math.exp(((x - mi) / (sigma * (math.sqrt(2))))**2)

b = (sigma * (math.sqrt(2*math.pi)))

resolucao = a/b
print(resolucao)
```

**Problema 2.** Vamos retomar o exercício sobre a população do Reino Unido trabalhado no exercício 2. Você já o resolveu utilizando Excel, agora implementaremos seu modelo em Python!

** População do Reino Unido **

<img src="Mapa_reino_unido.png"/, width=250, height=250 style="float: right; margin: 0px 0px 10px 10px">
O Reino Unido, situado em um conjunto de ilhas localizadas próximo à costa do continente europeu, é constituído por quatro países: Inglaterra, Escócia, País de Gales e Irlanda do Norte. Embora estejamos utilizando o termo “países”, cada um desses territórios não chega a ter a autonomia de um país, pois, para diversos assuntos, estão subordinados ao parlamento do Reino Unido, situado em Londres, capital do país. Para se ter uma ideia da confusão, cada um dos quatro disputa separadamente as principais competições de futebol: na Eurocopa de 1996, por exemplo, a Inglaterra derrotou a Escócia por 2 a 0. Porém, os quatro disputam os Jogos Olímpicos unidos sob a mesma bandeira, a do Reino Unido.



Confusões à parte, vamos construir um modelo da população do Reino Unido. Em 2008, sua população total era de cerca de 60 milhões de habitantes, sendo 50 milhões na Inglaterra, 5 milhões na Escócia, 3 milhões no País de Gales e 2 milhões na Irlanda do Norte. No mesmo ano, o governo instituiu uma nova política para a migração – enquanto os habitantes do Reino Unido eram autorizados a migrar livremente pelos quatro países, ninguém estava autorizado a sair ou entrar em seu território . Além disso, graças à paixão da maioria dos habitantes do Reino Unido pelos pubs e pelo futebol, o final de semana dos britânicos é cada vez mais atribulado, fazendo com que as taxas de natalidade e mortalidade tenham se equiparado, permanecendo assim por muitos anos. Os demógrafos (estudiosos das populações) estimam que, todos os anos, 10% da população escocesa migre para a Inglaterra, enquanto que 1% dos habitantes da Inglaterra se mudem para a Escócia. Além disso, 1000 habitantes do País de Gales mudam-se anualmente para a Inglaterra. Não é previsto que ocorra qualquer movimento populacional envolvendo a Irlanda do Norte.

(b)	Escreva equações a diferenças para modelar essas populações.

(c) Gere tres graficos, para as populações da Inglaterra, da Escocia, e um para Irlanda.
Não esqueça de nomear eixos.

```python
import matplotlib.pyplot as plt
from pylab import *

%matplotlib inline

#tempo em anos
T = 50

#parametros
Me = 0.1
Mi = 0.01
Mg = 1000

#Estoques
pop_ing = 50000000
pop_esc = 5000000
pop_gales = 3000000

# inicializa o vetor da série temporal
Ing = [pop_ing]
Esc = [pop_esc]
Gales = [pop_gales]

#rodando
for i in range (1,T):
    DeltaIng = Me * Esc[i -1] - Mi * Ing[i -1] + Mg
    DeltaEsc = Mi * Ing [i -1] - Me * Esc [i -1]
    DeltaGales = - Mg
    
    Ing.append(Ing[i -1] + DeltaIng)
    Esc.append(Esc[i - 1] + DeltaEsc)
    Gales.append(Gales[i -1] + DeltaGales)
tempo = range(0,T)
plt.plot(tempo, Ing )
plt.axis([0, T, min(Ing), max(Ing)])
plt.ylabel('população [Inglaterra]')
plt.xlabel('tempo [anos]')
title(r'População da Inglaterra')
plt.show()

tempo = range(0,T)
plt.plot(tempo, Esc )
plt.axis([0, T, 5000000, 5002000])
plt.ylabel('população [Escócia]')
plt.xlabel('tempo [anos]')
title(r'População da Escócia')
plt.show()

tempo = range(0,T)
plt.plot(tempo, Gales )
plt.axis([0, T, 2950000, 3000000])
plt.ylabel('população [Gales]')
plt.xlabel('tempo [anos]')
title(r'População de Gales')
plt.show()
```
