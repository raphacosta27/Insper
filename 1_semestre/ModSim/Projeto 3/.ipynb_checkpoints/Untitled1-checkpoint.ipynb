```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def derivasRapha (Y, t):
    
    #dEdT = ingA*Ka-KeE*Y[0]
    if t > 0.5 and t< 1:        
        dEdT = ingA*kp - KeE*Y[0] - KmE * Y[0] - KcM*Y[0] - KaM*Y[0]
    elif t > 1.1 and t < 1.4:
        dEdT = ingA*kp - KeE*Y[0] - KmE * Y[0] - KcM*Y[0] - KaM*Y[0]
    elif t > 1.5 and t < 1.9:
        dEdT = ingA*kp - KeE*Y[0] - KmE * Y[0] - KcM*Y[0] - KaM*Y[0]
    elif t > 2.1 and t < 2.4:
        dEdT = ingA*kp - KeE*Y[0] - KmE * Y[0] - KcM*Y[0] - KaM*Y[0]
    elif t > 2.5 and t < 2.9:
        dEdT = ingA*kp - KeE*Y[0] - KmE * Y[0] - KcM*Y[0] - KaM*Y[0]
    elif t > 3.1 and t < 3.4:
        dEdT = ingA*kp - KeE*Y[0] - KmE * Y[0] - KcM*Y[0] - KaM*Y[0]
    elif t > 3.5 and t < 3.9:
        dEdT = ingA*kp - KeE*Y[0] - KmE * Y[0] - KcM*Y[0] - KaM*Y[0]
    elif t > 4.1 and t < 4.4:
        dEdT = ingA*kp - KeE*Y[0] - KmE * Y[0] - KcM*Y[0] - KaM*Y[0]
    elif t > 4.5 and t < 4.9:
        dEdT = ingA*kp - KeE*Y[0] - KmE * Y[0] - KcM*Y[0] - KaM*Y[0]
    elif t > 5.1 and t < 5.4:
        dEdT = ingA*kp - KeE*Y[0] - KmE * Y[0] - KcM*Y[0] - KaM*Y[0]
    elif t > 5.5 and t < 5.9:
        dEdT = ingA*kp - KeE*Y[0] - KmE * Y[0] - KcM*Y[0] - KaM*Y[0]
    else:
        dEdT = - KeE*Y[0] - KmE * Y[0] - KcM*Y[0] - KaM*Y[0]
    
        
    dSdT = KmE * Y[0] - KeS*Y[1]
    return dEdT, dSdT
```

```python
ingA = 8.4 #Quantidade de álcool ingerido em g/h
kp = 1 #Constante de conversão em 1/L
KeE = 0.4 #Eliminação de álcool do estômago
KmE = 0.5 #Constante de metabolização do estômago
KcM = 0 #Constante de como a comida afesta a metabolização
KaM = 0 #Constante de como a água afesta a metabolização
KeS = 0.7 #Eliminação de álcool do sangue

tempo = np.linspace (0, 6, 100)

Eo = 1.8
So = 0

Y0 = [Eo, So]

teste = odeint(derivasRapha, Y0, tempo)




plt.plot(tempo, teste[:,0])
plt.ylabel("Concentração (g/L)")
plt.xlabel("Tempo (h)")
plt.grid(True)
plt.show()

plt.plot(tempo, teste[:,1])
plt.ylabel("Concentração (g/L)")
plt.xlabel("Tempo (h)")
plt.grid(True)
plt.show()
```
