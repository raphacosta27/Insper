```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def derivasRapha (Y, t):
    
    #dEdT = ingA*Ka-KeE*Y[0]
    if t > 0.5 and t< 1:        
        dEdT = ingA - KeE*Y[0] - KmE * Y[0] - Kcm*Y[0] - Kca*Y[0]
    elif t > 1.1 and t < 1.4:
        dEdT = ingA - KeE*Y[0] - KmE * Y[0] - Kcm*Y[0] - Kca*Y[0]
    elif t > 1.5 and t < 1.9:
        dEdT = ingA - KeE*Y[0] - KmE * Y[0] - Kcm*Y[0] - Kca*Y[0]
    elif t > 2.1 and t < 2.4:
        dEdT = ingA - KeE*Y[0] - KmE * Y[0] - Kcm*Y[0] - Kca*Y[0]
    elif t > 2.5 and t < 2.9:
        dEdT = ingA - KeE*Y[0] - KmE * Y[0] - Kcm*Y[0] - Kca*Y[0]
    elif t > 3.1 and t < 3.4:
        dEdT = ingA - KeE*Y[0] - KmE * Y[0] - Kcm*Y[0] - Kca*Y[0]
    elif t > 3.5 and t < 3.9:
        dEdT = ingA - KeE*Y[0] - KmE * Y[0] - Kcm*Y[0] - Kca*Y[0]
    elif t > 4.1 and t < 4.4:
        dEdT = ingA - KeE*Y[0] - KmE * Y[0] - Kcm*Y[0] - Kca*Y[0]
    elif t > 4.5 and t < 4.9:
        dEdT = ingA - KeE*Y[0] - KmE * Y[0] - Kcm*Y[0] - Kca*Y[0]
    elif t > 5.1 and t < 5.4:
        dEdT = ingA - KeE*Y[0] - KmE * Y[0] - Kcm*Y[0] - Kca*Y[0]
    elif t > 5.5 and t < 5.9:
        dEdT = ingA - KeE*Y[0] - KmE * Y[0] - Kcm*Y[0] - Kca*Y[0]
    else:
        dEdT = - KeE*Y[0] - KmE * Y[0] - Kcm*Y[0] - Kca*Y[0]
    
        
    dSdT = KmE * Y[0] - KeS*Y[1]
    return dEdT, dSdT
```
